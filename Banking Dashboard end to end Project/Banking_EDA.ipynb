### Univariate Analysis : Univariate analysis is the statistical examination of a single variable to understand its distribution, central tendency, and spread without considering relationships with other variables.


```python
# Univariate analysis is the process of examining and summarizing a single variable to understand its distribution, central tendency, and spread.

# List of categorical predictors to plot
predictors = ["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", 
              "Occupation", "Fee Structure", "Loyalty Classification", 
              "Properties Owned", "Risk Weighting", "Income Band"]

# Loop through each predictor in the list for univariate count plots
for i, predictor in enumerate(df_banking[predictors].columns):
    
    # Create a new figure for each plot with appropriate size
    plt.figure(i, figsize=(10, 5))

    # Create a univariate count plot (no hue)
    sns.countplot(data=df_banking, x=predictor, palette='Set2')

    # Set the title of the plot
    plt.title(f'Distribution of {predictor}')

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45)

    # Adjust layout to avoid overlapping
    plt.tight_layout()

    # Show the plot
    plt.show()
```

### Bivariate Analysis: Bivariate analysis is the statistical examination of the relationship between two variables to understand how one variable influences or correlates with the other

```python
# Bivariate analysis involves exploring the relationship between two variables to understand how one variable may be influenced by or correlated with the other.

# List of categorical predictors to plot
predictors = ["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", 
              "Occupation", "Fee Structure", "Loyalty Classification", 
              "Properties Owned", "Risk Weighting", "Income Band"]

# Loop through each predictor in the list
for i, predictor in enumerate(df_banking[predictors].columns):
    
    # Skip plotting if the predictor is the same as the hue (GenderId)
    if predictor == 'GenderId':
        continue

    # Create a new figure for each plot with appropriate size
    plt.figure(i, figsize=(10, 5))

    # Create a count plot with GenderId as hue for color separation
    sns.countplot(data=df_banking, x=predictor, hue='GenderId', palette='Set2')  # GenderId (1 = Male, 2 = Female)

    # Set the title of the plot
    plt.title(f'{predictor} by Gender')

    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45)

    # Adjust layout to avoid overlapping
    plt.tight_layout()

    # Show the plot
    plt.show()
```

### Numerical Analysis: Numerical analysis is the examination of numerical (quantitative) data to understand patterns, trends, distributions, and relationships between numeric variables.

```python
# Define the list of numerical columns for univariate analysis
numerical_cols = [
    'Estimated Income', 
    'Superannuation Savings', 
    'Credit Card Balance', 
    'Bank Loans', 
    'Bank Deposits', 
    'Checking Accounts', 
    'Saving Accounts', 
    'Foreign Currency Account', 
    'Business Lending'
]

# Create a large figure to hold multiple subplots (4 rows × 3 columns)
plt.figure(figsize=(15, 10))

# Loop through each numerical column and plot its distribution
for i, col in enumerate(numerical_cols):
    
    # Define subplot position (i+1 since subplot index starts at 1)
    plt.subplot(4, 3, i + 1)

    # Plot histogram with KDE using your DataFrame df_banking
    sns.histplot(df_banking[col], kde=True)

    # Set the title for the current subplot
    plt.title(col)

# Adjust layout to avoid overlapping of subplots
plt.tight_layout()

# Display the full set of histograms
plt.show()
```

```python
# Plot a histogram to visualize the distribution of estimated income

# Create a new figure with specified size (width=10 inches, height=5 inches)
plt.figure(figsize=(10, 5))

# Create a histogram for the 'Estimated Income' column
# - bins=20 divides the data into 20 intervals
# - kde=True overlays a smooth Kernel Density Estimate curve
# - color='skyblue' sets the fill color of the bars
sns.histplot(data=df_banking, x='Estimated Income', bins=20, kde=True, color='skyblue')

# Set the title of the plot
plt.title('Histogram of Estimated Income')

# Label the x-axis and y-axis
plt.xlabel('Estimated Income')
plt.ylabel('Frequency')

# Adjust layout to prevent overlap of plot elements
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Plot count of each occupation in the dataset

# Create a new figure with a specified size (width=12 inches, height=6 inches)
plt.figure(figsize=(12, 6))

# Create a count plot for the 'Occupation' column using a Set2 color palette
# This automatically counts how many times each occupation appears
sns.countplot(data=df_banking, x='Occupation')

# Set the title of the plot
plt.title('Count Plot of Occupation')

# Label the x-axis and y-axis
plt.xlabel('Occupation')
plt.ylabel('Count')

# Rotate x-axis labels by 45 degrees for better readability
plt.xticks(rotation=90)

# Adjust the layout to ensure labels and elements don't overlap
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Get the top 15 most frequent occupations
top_occupations = df_banking['Occupation'].value_counts().nlargest(15).index

# Filter the DataFrame to only include those occupations
filtered_df = df_banking[df_banking['Occupation'].isin(top_occupations)]

# Plot count of top 15 occupations
plt.figure(figsize=(12, 6))
sns.countplot(data=filtered_df, x='Occupation', palette='Set2',
              order=top_occupations)  # Ensures correct order

# Add titles and labels
plt.title('Count Plot of Top 15 Occupations')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Optional: Show distribution of estimated income by occupation using a boxplot

# Create a new figure with a specified size (14 inches wide by 6 inches tall)
plt.figure(figsize=(14, 6))

# Create a boxplot to visualize the distribution of 'Estimated Income' across different 'Occupation' categories
# - This shows median, interquartile range, and outliers for each occupation
# - 'palette=Set3' sets a colorful predefined color palette
sns.boxplot(data=df_banking, x='Occupation', y='Estimated Income', palette='Set3')

# Set the title of the plot
plt.title('Estimated Income by Occupation')

# Label the x-axis and y-axis
plt.xlabel('Occupation')
plt.ylabel('Estimated Income')

# Rotate the x-axis labels by 90 degrees for better readability
plt.xticks(rotation=90)

# Adjust layout to prevent overlap of labels and ensure everything fits
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Get top 15 most frequent occupations based on their occurrence in the dataset
top_occupations = df_banking['Occupation'].value_counts().nlargest(15).index

# Filter the DataFrame to only include rows with one of the top 15 occupations
filtered_df = df_banking[df_banking['Occupation'].isin(top_occupations)]

# Create a new figure for the boxplot with specified dimensions
plt.figure(figsize=(14, 6))

# Create a boxplot to show the distribution of 'Estimated Income' for each of the top 15 occupations
# This helps compare median, spread, and outliers across different occupations
sns.boxplot(data=filtered_df, x='Occupation', y='Estimated Income', palette='Set3')

# Set the title and axis labels
plt.title('Estimated Income by Top 15 Occupations')
plt.xlabel('Occupation')
plt.ylabel('Estimated Income')

# Rotate x-axis labels by 90 degrees for better readability
plt.xticks(rotation=90)

# Adjust layout to prevent overlapping elements
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Define the list of numerical columns to include in the correlation analysis
numerical_cols = [
    'Estimated Income', 
    'Superannuation Savings', 
    'Credit Card Balance', 
    'Bank Loans', 
    'Bank Deposits', 
    'Checking Accounts', 
    'Saving Accounts', 
    'Foreign Currency Account', 
    'Business Lending'
]

# Calculate the correlation matrix for the selected numerical columns
correlation_matrix = df_banking[numerical_cols].corr()

# Set the size of the heatmap figure
plt.figure(figsize=(12, 12))

# Create a heatmap to visualize the correlation matrix
# - annot=True displays correlation values in each cell
# - cmap='crest' sets the color scheme
# - fmt=".2f" formats the correlation values to 2 decimal places
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")

# Set the title of the heatmap
plt.title("Correlation Matrix of Numerical Features")

# Display the heatmap
plt.show()
```
